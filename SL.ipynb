```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import scipy.optimize
%pylab inline --no-import-all
Mlog=[0,-0.03,-0.11,-0.16,-0.33,-0.41,-0.67,-1.0,-10**9,-10**9,-10**10]
Llog=[0,-0.1,-0.4,-0.8,-1.2,-1.5,-2.1,-3.1,-10**5,-10**6,-10**7]
#Dataload
M=np.exp(Mlog)
L=np.exp(Llog)
print(M)
print(L)
X=np.linspace(1,0.1,10)#Mass distribution
print(X)
parameter_initial = np.array([0,0])

#fittinf func
def func(x,a,b):
    return  b * np.exp(x)+a
j=4
paramater_optimal1, covariance = scipy.optimize.curve_fit(func, M[:j], L[:j], p0=parameter_initial)
paramater_optimal2, covariance = scipy.optimize.curve_fit(func, M[j:], L[j:], p0=parameter_initial)
print ("paramater =", paramater_optimal1)
print ("paramater =", paramater_optimal2)
#Luminocity fitting
y1 = func(X,paramater_optimal1[0],paramater_optimal1[1])
y2 = func(X,paramater_optimal2[0],paramater_optimal2[1])
#plt.xscale('log')
#plt.yscale('log')
plt.plot(M, L, 'o')
plt.plot(X, y1, '-')
plt.plot(X, y2, '-')
plt.ylim(-0.1, 1.1)
plt.show()
k=3 #cutoff from last mass
#print(X)
print(y1)
print(y1[0]/y1[:k])
print(y1[0]/y2[k:])
Y1=y1[0]/y1[:k]
Y2=y1[0]/y2[k:]
rrnorm=np.concatenate([Y1,Y2])#行列結合
print(rrnorm)
rnorm =rrnorm[::-1]/2.7
print (rnorm)
print (1/rnorm)
Xr=X[::-1]
#plt.xscale('log')
#plt.yscale('log')
plt.plot(Xr, 1/rnorm, '-')

plt.show()
```

```python
plt.plot(Xr, 1/rnorm, '-')
plt.plot(Xr, 2.7/Xr, '-')
plt.show()
```
