```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import time
start = time.time() 
# self id mass disk 
#depe haba
num=22
denum=20
habnum=30
yrnum=3000
dlifes=np.arange(20)*10
#for starnum in [220000,100000]:
for starnum in [1000,5000,10000,100000]:
    for dlife in  dlifes:
        depes=np.arange(denum)/-50.#0.05 0.1 0.15 0.2
        habas=np.arange(habnum)/10. #0.1 0.3 0.5
        fulldata=np.zeros([denum,habnum,num,yrnum])
        print(depes,habas)
        for depenum in range(denum):###
            depe=depes[depenum]
            print(depe)
            for habanum in range(habnum):###
                haba=habas[habanum]
                print(haba)
                total=starnum
                stars=[]
                #taulife0=10
                taulife0=dlife#Myr
                ###depe=-0.05#dependence mass life
                #depe=-0.5 #dependence mass life
                ###haba=0.5
                num=22
                Ms=(np.arange(num)/(num-2))+0.05#0.05~1.1

                class star:
                    def __init__(self):                  # コンストラクタ
                        self.name = ""
                    def setid(self,numb):             # setName()メソッド
                        self.id = numb
                        self.disk=1
                    def setmass(self,total=starnum):
                        #self.mass=int(np.floor(self.id*10/total)) +1
                        #self.mass=np.random.choice(Ms, 1, replace=True)
                        stmassnumber=int(np.floor(self.id*num/total))   
                        self.mnum=stmassnumber  
                        self.mass=Ms[stmassnumber]
                    def setlifetime(self):
                        #self.lifetime=taulife0*((self.mass/10.)**depe)*(np.max(0,1+((np.random.rand()-0.5))*haba))/1.
                        life0 = np.array([0,1+((np.random.rand()-0.5))*haba])
                        self.lifetime=taulife0*((self.mass/10.)**depe)*(np.max(life0))*haba/1.
                    def diskdis(self):
                        self.disk=0        
                #star make

                #makestar
                for i in range (starnum):
                    s=star()
                    s.setid(i)
                    s.setmass()
                    s.setlifetime()
                    stars.append(s)
                #print(len(stars))
                #disk evo
                totaltime=30
                evo=np.zeros([num,yrnum])
                for t in range (totaltime*10):
                    #print(t)
                    msss=np.zeros(num)#disk　less　star　count
                    for st in  stars:
                        if st.lifetime<t/10.:
                            st.diskdis()#diskdispersion
                        for i in range(num):
                            if st.mnum==i:
                                msss[i]+=st.disk
                        #if st.id in [1,10,100,1000]:
                            #print(st.id,st.mass,st.lifetime)

                    for mss in range(num):
                        evo[mss,t]=msss[mss]/(starnum/num)
                #print(evo)
                fulldata[depenum,habanum,:,:]=evo



                # 1. 必要なモジュールの読み込み
                %matplotlib notebook
                from matplotlib.animation import ArtistAnimation

                # 2.グラフ領域の作成
                '''
                fig, ax = plt.subplots()
                plt.xlim(-0.1, 1.15)
                plt.ylim(-0.01, 1.01)
                #plt.text(0.8, 0.8, str(i/10)+"Myr", fontsize="xx-large")

                masses=(np.arange(10)+1)/10.
                # 3. グラフ要素のリスト（artists）作成
                artists = []
                for i in range(300):
                    #artist = ax.plot(masses,evo[:,i]/100, color='blue')
                    artist = ax.plot(Ms,evo[:,i]/(starnum/num), color='blue')
                    #artist =ax.text(0.8, 0.8, str(i/10)+"Myr")
                    artists.append(artist)

                # 4. アニメーション化
                anim = ArtistAnimation(fig, artists, interval=30)
                plt.show()'''

                #anim.save('005_50.mp4')
        #print(fulldata)
        end = time.time()  # 現在時刻（処理完了後）を取得
        time_diff = end - start  # 処理完了後の時刻から処理開始前の時刻を減算する
        print(time_diff)  # 処理にかかった時間データを使用

        np.save('bayo14lifes'+str(dlife)+'life'+str(starnum)+'.npy', fulldata)
```
