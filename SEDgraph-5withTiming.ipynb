```python
%matplotlib nbagg
import matplotlib.animation as animation

masss=np.arange(0.1,1.01,0.01)
massi=masss*10
massi=massi.astype(np.int64)*0.1
massi0=massi*10-1
massi0=massi0.astype(np.int64)
massi1=massi*10
massi1=massi1.astype(np.int64)
zetakro1=masss**-2.35
zetakro2=(masss**-1.35)*2
zetakro=np.append(zetakro2[:40],zetakro1[40:])

for ini in range(8):
    avetime=5
    fig = plt.figure()
    def updata(i):
        if i != 0:
            plt.cla()
            speed=i  #speedup speed=i*50
            slopes=lall[ini,i,:]
            slopess=np.zeros(91) #each starmass s @time
            allslps=np.zeros((300,91))
            for m in range(91):
                slopess[m]=slopes[massi0[m]]*(1-((masss[m]-massi[m])*10))+slopes[massi1[m]]*((masss[m]-massi[m])*10)
            allslps[i,:]=slopess
            print(allslps)
            #plt.plot(slopess,zetakro)
            slpss=np.zeros(91)
            if i>avetime:
                for ave in range(avetime-1):
                    slpss+=allslps[i-ave,:]
                    #print(slpss[0],allslps[i-ave,0],ave,i-ave)
                plt.semilogy(slpss,zetakro,'o')
                plt.xlim(-3, 1)
            else:
                plt.semilogy(slopess,zetakro,'o')
                plt.xlim(-3, 1)
    ani = animation.FuncAnimation(fig, updata, interval=100, frames=300)
    ani.save("animeSEDslope"+str(avetime)+"ave"+str(ini)+"0624.mp4", writer="ffmpeg")
```
