```python
wavobs=(0.44e-23*3e22/(wav))* u.AA*fluxtest.unit# L=M^2
lifetime=np.ones((30, 10,8))

import time
import datetime

start = time.time()

pltdat=dataall[:,:,1]
x =pltdat[:,0]
for inicon in range(6):
    inicon=inicon+2  #restart +? 8-?
    ini=(inicon+1)*10
    print(ini,"   ",datetime.datetime.now())
    for i in range(300):  #time to300 or short test
        for j in range(10): #10=0.1~1.0 1=1mass
            if i%100==0:
                t = time.time() - start
                print(ini,"   ",i,"   ",j,"   ",t,"   ",datetime.datetime.now())
            ininum=ini-9+j
            pltdat=dataall[:,:,ininum]
            s =pltdat[:,i+1]                          
            s[s<1e-8]=0 #kagen ika ha 0 ni suru   
            #s[s<1e-6]=0 #kagen ika ha 0 ni suru
            t =pltdat[:,i+301]             
            m=0
            l=0
            fluxall = 0
            while (t[l]>limKlow):
                l=m
                flux = blackbody_lambda(wav, t[l]*u.K)
                kei=(1-np.exp(-(10*s[l])))*x[l]*(x[l+1]-x[l])
                F=flux*kei*pis
                fluxall+=F
                m+=1
            if i>2:
                for k in obs:
                    if fluxall[k]*wav[k]>wavobs[k]:
                        lifetime[k,j,inicon]=i
print(lifetime)
```
